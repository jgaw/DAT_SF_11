```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('../data/wine.csv')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scale = StandardScaler()
pca = PCA(n_components=12)


X = df.ix[:,1:]
X_scaled = scale.fit_transform(X)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10 , max_iter = 300, random_state=1)
Y_hat_kmeans = kmeans.fit(X_pca).labels_

```

```python
mu = kmeans.cluster_centers_
plt.scatter(X_pca[:,0], X_pca[:,1], c=Y_hat_kmeans, alpha=0.4)
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat_kmeans))
print mu
```

```python
# compute distance matrix
from scipy.spatial.distance import pdist, squareform
distx = squareform(pdist(X_pca, metric='euclidean'))
distx
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

R = dendrogram(linkage(distx, method='ward'), color_threshold=100)

plt.xlabel('points')
plt.ylabel('Height')
plt.suptitle('Cluster Dendrogram', fontweight='bold', fontsize=14);
```
